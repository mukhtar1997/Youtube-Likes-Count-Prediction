```python
h=widgets.Dropdown(
    options=video['channel_title'].unique(),
    value='Good Mythical Morning',
    description='Channel name:',
    disabled=False,
)
def piegraph(ch):
    nbc=video[video['channel_title']==ch]
    lk=nbc['likes'].sum()
    dk=nbc['dislikes'].sum()
    cc=nbc['comment_count'].sum()
    plt.pie(x=[lk,dk,cc],labels=['likes','dislikes','comment'],autopct='%1.1f%%')
    plt.show()
```

```python
widgets.interact(piegraph,ch=h)
```
